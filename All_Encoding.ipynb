```python
import numpy as np
import matplotlib.pyplot as plt

#Unipolar Encoding
data_input=input()
data=[int(i) for i in data_input]
Time=np.arange(len(data)+1)
signal_1=np.zeros(len(data)+1,dtype=int)
for i in range(len(data)):
    if data[i]==0:
        signal_1[i]=0
    else:
        signal_1[i]=1

#plotting
plt.subplot(3,2,1)
plt.title("UNIPOLAR ENCODDING")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(Time,signal_1,where="post")
plt.text(0,2,data)
plt.grid(True)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)

#plt.show()      


#NRZ-L Encoding
if data[0] == 0:
    nrz_data = [1]
else:
    nrz_data = [-1]

for i in range(len(data)):
    if data[i] == 0:
        nrz_data.append(1)
    else:
        nrz_data.append(-1)

Time = np.arange(len(nrz_data))

#ploting 
plt.subplot(3,2,2)

plt.title("NRZ-L Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(Time, nrz_data)
plt.yticks([-3, -2, -1, 0, 1, 2, 3])
plt.xticks(Time)
plt.grid(True)
plt.axhline(y=0, color="green")
#plt.show()


#NRZ-I Encoding
bit=1
signal_list=[1]

for i in range(0,len(data)):
    if data[i]==1:
        if bit==1:
            bit=-1
        else:
            bit=1
    signal_list.append(bit)
#print(signal_list)

#ploting
plt.subplot(3,2,3)
Time=np.arange(0,len(signal_list))

plt.title("NRZ-I Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(Time,signal_list)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.axhline(y=0,color="green")
plt.grid(True)

#RZ Encoding
data_1=np.arange(0,len(data_input),0.5)
signal=np.array([])
for i in data_input:
    if i=='1':
        signal=np.append(signal,[1,0])
    else:
        signal=np.append(signal,[-1,0])

Time=np.arange(0,len(data_input))

#ploting
plt.subplot(3,2,4)
plt.title("RZ Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data_1,signal,where="post")
plt.xticks(Time)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.grid(True)


#Manchester Encoding

data_1=np.arange(0,len(data_input),0.5)
signal=np.array([])

for i in data_input:
    if i=='1':
        signal=np.append(signal,[-1,1])
    else:
        signal=np.append(signal,[1,-1])
        
#ploting
plt.subplot(3,2,5)
Time=np.arange(0,len(data_input))     

plt.title("Manchester ENcoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data_1,signal,where="post")
plt.xticks(Time)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.grid(True)
#plt.show()


#Differential Manchester
data_1=np.arange(0,len(data_input),0.5)
signal=np.array([])


temp=1
for i in data_input:
    if i=='1':
        signal=np.append(signal,[temp,temp*-1])
        temp=temp*-1
    else:
        signal=np.append(signal,[temp*-1,temp])
    
Time=np.arange(0,len(data_input))

#ploting
plt.subplot(3,2,6)
Time=np.arange(0,len(data_input))

plt.title("Differemtial Manchester")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data_1,signal,where="post")
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.grid(True)

plt.tight_layout()
plt.show()
```
