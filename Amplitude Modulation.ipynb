```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
carrier_Amplitude=4
carrier_Frequency=3
Time=np.arange(0,2*np.pi,0.001)
carrier_signal=carrier_Amplitude*np.sin(2*carrier_Frequency*np.pi*Time)
```

```python
Modulatin_Amplitude=2
Modulating_Frequency=0.5
Modulating_signal=Modulatin_Amplitude*np.sin(2*Modulating_Frequency*np.pi*Time)
```

```python
Modulated_signal=carrier_signal*Modulating_signal
```

```python

plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.grid(True)
plt.plot(Time,carrier_signal)
plt.yticks([-4,0,4])




plt.subplot(3,1,2)
plt.title("Modulating Signal")
plt.grid(True)
plt.plot(Time,Modulating_signal)
plt.yticks([-2,0,2])





plt.subplot(3,1,3)
plt.title("Amplitude Modulated Signal")
plt.grid(True)
plt.plot(Time,Modulated_signal)
```
