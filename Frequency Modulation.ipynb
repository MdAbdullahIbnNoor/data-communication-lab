```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
carrier_Amplitude=1
carrier_Frequency=3
Time=np.arange(0,2*np.pi,0.001)
carrier_signal=carrier_Amplitude*np.sin(2*carrier_Frequency*np.pi*Time)
```

```python
message_Amplitude=2
message_Frequency=0.5
message_signal=message_Amplitude*np.sin(2*message_Frequency*np.pi*Time)
```

```python
Modulated_signal=carrier_Amplitude*np.sin(2*np.pi*((message_Frequency*Time)+(2*message_signal)))
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(Time,carrier_signal)
plt.grid(True)


plt.subplot(3,1,2)
plt.title("Message Signal")
plt.plot(Time,message_signal)
plt.grid(True)


plt.subplot(3,1,2)
plt.title("Frequency Modulated  Signal")
plt.plot(Time,Modulated_signal)
plt.grid(True)
plt.show()

```
