```python
#Unipolar Encoding

import numpy as np
import matplotlib.pyplot as plt

data_input=input()
data=[int(i) for i in data_input]


Time=np.arange(len(data)+1)
signal_1=np.zeros(len(data)+1,dtype=int)

for i in range(len(data)):
    if data[i]==0:
        signal_1[i]=0
    else:
        signal_1[i]=1
        


plt.title("Unipolar Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(Time,signal_1,where="post")
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
#plt.axhline(y=0,color="green")
plt.text(1,2,data)
plt.grid(True)

plt.show()



```

```python
#NRZ-L encoding
import numpy as np
import matplotlib.pyplot as plt

data_input=input()
data=[int(i) for i in data_input]

if data[0]==0:
    nrz_data=[1]
else:
    nrz_data=[-1]
    
for i in range(len(data)):
    if data[i]==0:
        nrz_data.append(1)
    else:
        nrz_data.append(-1)

Time=np.arange(len(nrz_data))


plt.title("NRZ-L Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(Time,nrz_data)
plt.axhline(y=0,color="green")
plt.xticks(Time)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.text(1,2,data)
plt.grid(True)
plt.show()
```

```python
#NRZ-I Encoding
import numpy as np
import matplotlib.pyplot as plt

data_input=input()
data=[int(bit) for bit in data_input]


print(data)

bit=1
signal_list=[1]

for i in range(0,len(data)):
    if data[i]==1:
        if bit==1:
            bit=-1
        else:
            bit=1
    signal_list.append(bit)
print(signal_list)


Time=np.arange(0,len(signal_list))

plt.title("NRZ-I Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(Time,signal_list)
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.axhline(y=0,color="green")
plt.grid(True)

```

```python
#RZ Encoding
import numpy as np
import matplotlib.pyplot as plt

data_input=input()


data=np.arange(0,len(data_input),0.5)
signal=np.array([])

for i in data_input:
    if i=='1':
        signal=np.append(signal,[1,0])
    else:
        signal=np.append(signal,[-1,0])
        
Time=np.arange(0,len(data_input))

plt.title("RZ Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data,signal,where="post")
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.grid(True)
plt.show()
```

```python
#Manchester 

import numpy as np
import matplotlib.pyplot as plt

data_input=input()
print(data_input)

data=np.arange(0,len(data_input),0.5)
signal=np.array([])

for i in data_input:
    if i=='1':
        signal=np.append(signal,[-1,1])
    else:
        signal=np.append(signal,[1,-1])
        
Time=np.arange(0,len(data_input))

plt.title("Manchester Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data,signal,where="post")
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data_input=input()
print(data_input)

data=np.arange(0,len(data_input),0.5)
signal=np.array([])

temp=1
for i in data_input:
    if i=='1':
        signal=np.append(signal,[temp,temp*-1])
        temp=temp*-1
    else:
        signal=np.append(signal,[temp*-1,temp])
        
Time=np.arange(0,len(data_input))


plt.title("Differential Manchester Encoding")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data,signal,where="post")
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.grid(True)
plt.show()
```

```python
#AM


import numpy as np
import matplotlib.pyplot as plt

# carrier signal
ac=4
fc=3
t=np.arange(0,2*np.pi,0.001)
cs=ac*np.sin(2*fc*np.pi*t)

#Modulating signal
am=2
fm=0.5
ms=am*np.sin(2*fm*np.pi*t)

#Amplitude Modulation
A_M=cs*ms


plt.subplot(3,1,1)
plt.plot(t,cs)
plt.grid(True)
plt.yticks([-4,0,4])


plt.subplot(3,1,2)
plt.plot(t,ms)
plt.grid(True)
plt.yticks([-2,0,2])


plt.subplot(3,1,3)
plt.plot(t,A_M)
plt.grid(True)

```
