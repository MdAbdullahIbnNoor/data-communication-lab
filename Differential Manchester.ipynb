```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_input=input()
print(data_input)
```

```python
data=np.arange(0,len(data_input),0.5)
signal=np.array([])


temp=1
for i in data_input:
    if i=='1':
        signal=np.append(signal,[temp,temp*-1])
        temp=temp*-1
    else:
        signal=np.append(signal,[temp*-1,temp])
```

```python
Time=np.arange(0,len(data_input))

plt.title("Differemtial Manchester")
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.step(data,signal,where="post")
plt.yticks([-3,-2,-1,0,1,2,3])
plt.xticks(Time)
plt.grid(True)
plt.show()
```
