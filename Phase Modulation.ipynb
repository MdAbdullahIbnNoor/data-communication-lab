```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
carrier_Amplitude=1
carrier_Frequency=3

Time=np.arange(0,2*np.pi,0.001)

carrier_signal=carrier_Amplitude*np.sin(2*np.pi*carrier_Frequency*Time)
```

```python
Message_Amplitude=2
Message_Frequency=0.5

Message_signal=Message_Amplitude*np.sin(2*np.pi*Message_Frequency*Time)
```

```python
Modulated_signal=carrier_Amplitude*np.sin((2*np.pi*Message_Frequency*Time)+(2*Message_signal))
```

```python
plt.subplot(3,1,1)
plt.title("Carrier Signal")
plt.plot(Time,carrier_signal)
plt.grid(True)


plt.subplot(3,1,2)
plt.title("Modulating Signal")
plt.plot(Time,Message_signal)
plt.grid(True)


plt.subplot(3,1,3)
plt.title("Phase Modulated Signal")
plt.plot(Time,Modulated_signal)
plt.grid(True)
```
